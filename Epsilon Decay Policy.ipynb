```python
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
```

## Epsilon Anneal


```python
from tensorforce.core.explorations import EpsilonAnneal

policy = EpsilonAnneal(
    initial_epsilon=1.0, 
    final_epsilon=0.1, 
    timesteps=10000, 
    start_timestep=0
)

policy(episode=0, timestep=6000.) #Note the timestep must be foat for the policy to work properly
```

```python
x = np.arange(20000)
y = [policy(episode=0, timestep=float(i)) for i in x]
plt.plot(x,y)
```

## Linear Decay

```python
from tensorforce.core.explorations import LinearDecay
policy = LinearDecay()
policy(episode=10)
```

```python
x = np.arange(200)
y = [policy(episode=i) for i in x]
plt.plot(x,y)
```

## Epsilon Decay

```python
from tensorforce.core.explorations import EpsilonDecay
policy = EpsilonDecay(initial_epsilon=1.0, final_epsilon=0.1, timesteps=10000, start_timestep=0, half_lives=10)
policy(episode=0, timestep=6000.)
```

```python
x = np.arange(20000)
y = [policy(episode=0, timestep=float(i)) for i in x]
plt.plot(x,y)
```

## Ornstein Uhlenbeck Process

```python
from tensorforce.core.explorations import OrnsteinUhlenbeckProcess
policy = OrnsteinUhlenbeckProcess(sigma=0.3, mu=0, theta=0.15)
policy(episode=0, timestep=6000.)
```

```python
x = np.arange(20000)
y = [policy() for i in x]
plt.plot(x,y)
```
